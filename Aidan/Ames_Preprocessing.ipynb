```python
df = pd.read_csv('Ames_HousePrice.csv', index_col=[0])
df
```

```python
#remove unhelpful features...
#too many un-encodable NA's, feature is missing from too many homes to use in the remaining ones,
#houses overwhelmingly share the same characteristic/result for a specific feature, etc.

dropped_columns = ['PID','Alley', 'PoolArea', 'PoolQC', 'MiscFeature', 'MiscVal', 'Fence', 'FireplaceQu', 'LotFrontage', 'GarageYrBlt', 'Electrical', 'Functional', 'GarageQual', 'GarageCond', 'PavedDrive', 'SaleType', 'SaleCondition']
df = df.drop(columns=dropped_columns)
```

```python
#check the subsequent df again for NA's

df.info()
```

```python
# Where NA's are miniscule and can't be encoded, replace w mean or mode

df['MasVnrType']=df['MasVnrType'].fillna(df['MasVnrType'].mode()[0])
df['MasVnrArea']=df['MasVnrArea'].fillna(df['MasVnrArea'].mean())
```

```python
# almost all remaining NA's are intentional;
# used when house doesn't have a feature that's being graded / measured in that column.
# Fill with 0's

df = df.fillna(0)

df.info()
```

```python
#assign values to ORDINAL categorical variables
#doing this manually bc label_encoder doesn't seem to preserve correct order of values

ordinal_features_1 = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC', 'KitchenQual']

ordinal_values_1 = {'Po':1,'Fa':2,'TA':3,'Gd':4,'Ex':5}
cleanup_dict = {}

for i in ordinal_features_1:
    cleanup_dict[i] = ordinal_values_1
    

ordinal_values_2 = {'No':1, 'Mn':2, 'Av':3, 'Gd':4}
cleanup_dict['BsmtExposure'] = ordinal_values_2

ordinal_values_3 = {'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6}
cleanup_dict['BsmtFinType1'] = ordinal_values_3
cleanup_dict['BsmtFinType2'] = ordinal_values_3

ordinal_values_4 = {'Unf':1, 'RFn':2, 'Fin':3}
cleanup_dict['GarageFinish'] = ordinal_values_4

cleanup_dict
```

```python
df.replace(to_replace=cleanup_dict,value=None,inplace=True)
```
