```python
df = pd.read_csv('clean_ames_df.csv')
df
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn import datasets
```

```python
house_features = df.drop(columns=['SalePrice', 'MSSubClass'])
house_features
```

```python
prices = df['SalePrice']
prices
```

```python


ridge = Ridge()
```

```python
ridge.fit(house_features, prices)
print('the ridge intercept is: %.2f' %(ridge.intercept_))
pd.Series(ridge.coef_, index=house_features.columns)
```

```python
# Normalizing the data. 
# Choosing normalization over standardization...
# because EDA ssuggested that not all features' errors follow Gaussian distribution
```

```python
ridge.set_params(alpha = 1, normalize=True)
```

```python
ridge.fit(house_features, prices)
```

```python
print('the ridge intercept is: %.2f' %(ridge.intercept_))
pd.Series(ridge.coef_, index=house_features.columns)
```

```python
alphas = np.arange(0,20)
ridge.set_params(normalize=True)
coefs  = []
scores = []
for alpha in alphas:
        ridge.set_params(alpha=alpha)
        ridge.fit(house_features, prices)  
        coefs.append(ridge.coef_)
        scores.append(ridge.score(house_features, prices))
coefs = pd.DataFrame(coefs, index = alphas, columns = house_features.columns)  
coefs.head()
```

```python
#Let's retry ridge with cross-validation and tune hyperparameters
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold
from numpy import arange
```

```python
model = Ridge()
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
grid = dict()
grid['alpha'] = arange(0.1, 1.5, 0.01)
search = GridSearchCV(model, grid, scoring='r2', cv=cv, n_jobs=-1)
results = search.fit(house_features, prices)
print(r'The R^2 is %.2f' % results.best_score_)
print('Config: %s' % results.best_params_)
```

```python
ridge.set_params(alpha = 1, normalize=False)
ridge.fit(house_features, prices)
print('the ridge intercept is: %.2f' %(ridge.intercept_))
print(pd.Series(ridge.coef_, index=house_features.columns))
```

```python
# testing a wider range of alphas provides a "best parameter" of about 9.25 ...
# ... but that provides a negligible improvement in R^2 (about 0.1 better than alpha=1 ...
# ... while simultaneously raising the intercept above a reasonable expectation(almost 200k) 

# So I'm keeping alpha at 1
```

```python
# same as above but with normalization

######## Some of these coefficients make more intuitive sense than they did in non-normalized results

######## probably bc it improves relative accuracy of OHE'd features.

######## So I'd trust these results the most!

ridge.set_params(alpha = 1, normalize=True)
ridge.fit(house_features, prices)
print('the ridge intercept is: %.2f' %(ridge.intercept_))
print(pd.Series(ridge.coef_, index=house_features.columns))
```
